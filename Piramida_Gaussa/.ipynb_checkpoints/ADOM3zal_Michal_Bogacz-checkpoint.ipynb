```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# data import
location_1 = 'C:/Users/moko/OneDrive/Studia/MagisterkaINF/sem2/Wizja/LAB/ADOM/dowels.png'
location_2 = 'C:/Users/moko/OneDrive/Studia/MagisterkaINF/sem2/Wizja/LAB/ADOM/baboon.jpg'

image_1 = cv2.imread(location_1, 0)
image_2 = cv2.cvtColor(cv2.imread(location_2), cv2.COLOR_BGR2RGB)

# funkcja wyswietlająca obraz kolorowy lub w skali szarości
def pobr_cv2(obraz, tytul = '', osie = False, opencv = True):
    if not(osie):
        plt.axis("off") 
    if obraz.ndim == 2:
        plt.imshow(obraz,cmap='gray')
    else:
        if opencv:
            plt.imshow(obraz)
        else:
            plt.imshow(obraz)
    plt.title(tytul)

# funkcja wyświetlająca obrazy wg. listy
def pobr(listaobr, ile_k = 1, listatyt = [], opencv = True, wart_dpi = 100, osie = False):
    rozm_obr = 5
    ile = len(listaobr)
    if len(listatyt) == 0:
        listatyt = [' ']*ile
    ile_w = np.ceil(ile / ile_k).astype(int)
    figsize_k = rozm_obr*ile_k
    figsize_w = rozm_obr*ile_w
    plt.figure(figsize=(figsize_k,figsize_w), dpi = wart_dpi)
    for i in range(0,ile):
        if isinstance(listaobr[i],np.ndarray):
            plt.subplot(ile_w,ile_k,i+1)
            pobr_cv2(listaobr[i],listatyt[i],osie,opencv)
    plt.show()
```

```python
# kod funkcji
def gaussian_piramid(image, n_neighbours=3, step=2, std_range=(0,12), fitting_bottom=False):
    '''
    The designed function is supposed to return the created Gaussian pyramid in form of singular structure.
    The parameters:
    gaussian_piramid(image, n_neighbours=3, steo=2, std_range=(0,12))
    
    image - the image being filtered, either a gray scaled or any type of color scale
    n_neighbours - how many neighbours the filter mask should consider (mask_size = 2*neighbours_size + 1)
    step - the difference in std values between levels
    std_range - the range between which the base and the top layers are created
    fitting_bottom - the range values may not fit perfeclty to step values, so one may choos where to loose the 
        overfitting values, on the bottom or on the top of the piramid.
        Example: step=2, std range = (0,5)  - with fitting_bottom=False the layers are [0,2,4,5]
                                            - with fitting_bottom=True the layers are [0,1,3,5]
        This option is engaged only if the parameters are not dividable.
    '''
    std_min, std_max = std_range
    
    if type(fitting_bottom) != bool:
        raise Exception('Fitting_bottom should be True or False, for definitions check docstring')
    if type(std_range) != tuple: 
        raise Exception('The std_range given is not tuple')
    if std_min>std_max: 
        raise Exception('The std_range given is incorrect, firs param. is greater than last')
    if std_max-std_min < step: 
        raise Exception('The step value is greater than the diverence between max and min std value')
        
    # niepowtarzanie obrazu oryginalnego 
    if std_min==0: std_min=std_min + step    
        
    mask_size = n_neighbours*2 + 1
    
    if fitting_bottom == False:
        std_levels = np.arange(std_min, std_max+1, step, int)
        if std_max not in std_levels:
            std_levels = np.append(std_levels, std_max)
    else:
        std_levels = np.arange(std_max, std_min-1, -step, int)
        if std_min not in std_levels:
            std_levels = np.append(std_levels, std_min-1)
        std_levels = std_levels[::-1]
            
    lio = [image]
    lin = ['org, std=0']
    for i in std_levels:
        lio.append(cv2.GaussianBlur(image,(rozmiar_maski,rozmiar_maski),i))
        lin.append("std="+str(i))
        
    pobr(lio,3,lin)  
    return np.array(lio)
    
```
